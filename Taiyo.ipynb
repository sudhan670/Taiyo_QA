```python
import pandas as pd

# Create a list of dictionaries to represent the data
data = [
    {'entity_id': 1, 'entity_name': 'Tata Steel Ltd.', 'address': 'Mumbai'},
    {'entity_id': 2, 'entity_name': 'tata steel corporation', 'address': 'New Delhi'},
    {'entity_id': 3, 'entity_name': 'TATA STEEL LIMITED', 'address': 'Bangalore'},
    {'entity_id': 4, 'entity_name': 'Tata Steel', 'address': 'Mumbai'},
    {'entity_id': 5, 'entity_name': 'Tata Motors', 'address': 'Pune'},
    {'entity_id': 6, 'entity_name': 'Tata motors ltd.', 'address': 'Pune'},
    {'entity_id': 7, 'entity_name': 'TATA MOTORS LIMITED', 'address': 'Pune'},
    {'entity_id': 8, 'entity_name': 'Steel Authority of India', 'address': 'New Delhi'},
    {'entity_id': 9, 'entity_name': 'steel authority of india ltd.', 'address': 'Ranchi'},
    {'entity_id': 10, 'entity_name': 'SAIL', 'address': 'Kolkata'}
]

# Generate additional records by repeating and slightly modifying the base data
for i in range(51,101):
    entity_id = i
    if i % 10 == 1:
        entity_name = 'Tata Steel Ltd.'
        address = 'Mumbai'
    elif i % 10 == 2:
        entity_name = 'tata steel corporation'
        address = 'New Delhi'
    elif i % 10 == 3:
        entity_name = 'TATA STEEL LIMITED'
        address = 'Bangalore'
    elif i % 10 == 4:
        entity_name = 'Tata Steel'
        address = 'Mumbai'
    elif i % 10 == 5:
        entity_name = 'Tata Motors'
        address = 'Pune'
    elif i % 10 == 6:
        entity_name = 'Tata motors ltd.'
        address = 'Pune'
    elif i % 10 == 7:
        entity_name = 'TATA MOTORS LIMITED'
        address = 'Pune'
    elif i % 10 == 8:
        entity_name = 'Steel Authority of India'
        address = 'New Delhi'
    elif i % 10 == 9:
        entity_name = 'steel authority of india ltd.'
        address = 'Ranchi'
    else:
        entity_name = 'SAIL'
        address = 'Kolkata'

    data.append({'entity_id': entity_id, 'entity_name': entity_name, 'address': address})

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("sample_120_records.csv", index=False)

# Display the first few rows to verify
print(df.head(10))
print("\nTotal records:", len(df))
print(df)
```
